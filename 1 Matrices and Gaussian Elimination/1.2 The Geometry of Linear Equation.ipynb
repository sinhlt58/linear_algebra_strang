```python
# x  + y  = 4
# 2x - 2y = 4

x1 = np.linspace(0, 7)
y1 = 4 - x1

x2 = np.linspace(0, 7)
y2 = -2 + x2

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
